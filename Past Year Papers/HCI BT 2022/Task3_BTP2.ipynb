```python
# Task 3.1
class Node:
    def __init__(self, data):
        self.leftPtr = -1
        self.data = data
        self.rightPtr = -1

class BSTree:
    def __init__(self):
        self.thisTree = [Node(0),Node(0),Node(0),Node(0),Node(0),Node(0),Node(0),Node(0),Node(0),Node(0),Node(0)]
        self.root = 0
        self.nextFree = 1
        
        #constructor also known as init - should put in init
        self.thisTree[0].leftPtr = -1
        self.nextFree = 1
        for i in range(1,10):
            self.thisTree[i].leftPtr = i+1
        self.thisTree[10].leftPtr = -1
        
    def add(self, newItem):
        if self.nextFree == -1: #full
            print(f"Cannot insert '{newItem}'. Tree is full")
        elif self.nextFree == 1:
            self.nextFree = self.thisTree[1].leftPtr
            self.thisTree[0].leftPtr = 1
            self.thisTree[1].leftPtr = -1
            self.thisTree[1].data = newItem
#             print("root is inserted")
        else:
            ptr = self.thisTree[self.thisTree[self.root].leftPtr]
            while ptr:
#                 print("looping")
                if newItem < ptr.data:
                    if ptr.leftPtr == -1:
                        ptr.leftPtr = self.nextFree
                        self.nextFree = self.thisTree[self.nextFree].leftPtr
                        self.thisTree[ptr.leftPtr].leftPtr = -1
#                         self.thisTree[ptr.leftPtr].rightPtr = -1
                        self.thisTree[ptr.leftPtr].data = newItem
                        ptr = None
#                         print(f"{newItem} is inserted in a left branch")
                    else:
                        ptr = self.thisTree[ptr.leftPtr]
                elif newItem > ptr.data:
                    if ptr.rightPtr == -1:
                        ptr.rightPtr = self.nextFree
                        self.nextFree = self.thisTree[self.nextFree].leftPtr
                        self.thisTree[ptr.rightPtr].leftPtr = -1
                        self.thisTree[ptr.rightPtr].rightPtr = -1
                        self.thisTree[ptr.rightPtr].data = newItem
                        ptr = None
#                         print(f"{newItem} is inserted in a right branch")
                    else:
                        ptr = self.thisTree[ptr.rightPtr]
                else:
                    print(f"{data} already exists")
        
    def displayInorder(self, ptridx=None):
        if ptridx == None:
            ptridx = self.thisTree[self.root].leftPtr
            ptridx = 1
        if ptridx != -1:
            self.displayInorder(self.thisTree[ptridx].leftPtr)
            print(self.thisTree[ptridx].data)
            self.displayInorder(self.thisTree[ptridx].rightPtr)
        
    def displayPostOrder(self, ptridx = None):
        if ptridx == None:
            ptridx = self.thisTree[self.root].leftPtr
        if ptridx != -1:
            self.displayPostOrder(self.thisTree[ptridx].leftPtr) #you smh copy and pasted wrong
            self.displayPostOrder(self.thisTree[ptridx].rightPtr)
            print(self.thisTree[ptridx].data)

    def display(self): #supposed to display whole index, data, left, right thing
        print(f"index:{1}, data:{self.thisTree[1].data}, left:{self.thisTree[1].leftPtr}, right:{self.thisTree[1].rightPtr}")
        for i in range(2,11):
            print(f"index:{i}, data:{self.thisTree[i].data}, left:{self.thisTree[i].leftPtr}, right:{self.thisTree[i].rightPtr}")
```
