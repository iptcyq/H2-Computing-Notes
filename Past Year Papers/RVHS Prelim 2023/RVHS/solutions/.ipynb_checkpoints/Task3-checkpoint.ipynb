```python
# Task 3.1 [5]
from pymongo import MongoClient
import pymongo
import csv

def readfile(filename):
    results = []
    file = open(filename)
    csvreader = csv.reader(file)
    for row in csvreader:
        results.append(row)
    return results

def Task3_1():
    client = pymongo.MongoClient("127.0.0.1", 27017)
    client.drop_database("school")
    mydb = client['school']
    mycol = client['school']['teacher']
    
    results = readfile("teachers.csv")

    for i in range(1, len(results)):
        document = {} 
        if results[i][0] != "":
            document[results[0][0]] = int(results[i][0])
        if results[i][1] != "":
            document[results[0][1]] = results[i][1]
        if results[i][2] != "":
            document[results[0][2]] = results[i][2]
        if results[i][3] != "":
            document[results[0][3]] = results[i][3]
        if results[i][4] != "":
            document[results[0][4]] = results[i][4]
        if results[i][5] != "":
            document[results[0][5]] = int(results[i][5])
            
        # Insert data into database
        mycol.insert_one(document)
    client.close()


Task3_1()
```

```python
# Task 3.2 - display_all_teacher_details [3]
def Task_3_2(): 
    client = pymongo.MongoClient("127.0.0.1", 27017)
    mydb = client['school']
    mycol = client['school']['teacher']

    query = {}
    data = list(mycol.find(query))
    client.close()
    
    print("{:<5}{:<20}{:<15}{:<10}{:<20}{:<7}".format("_id","Name","Department","Contact","Role","Ratings"))
    for row in data:
        print("{:<5}{:<20}{:<15}{:<10}{:<20}{:<7}".format(row["_id"],
                                                          row["Name"],
                                                          row["Department"],
                                                          row["Contact"] if ("Contact" in row) else "-",
                                                          row["Role"] if ("Role" in row) else "-",
                                                          row["Ratings"] if ("Ratings" in row) else "-"))


Task_3_2()
```

```python
# Task 3.3 [2]
# display the names of teachers who Either 1) do not have both Contact and Role OR 
#                                          2) do not have a Role and is in the Math Department

def Task_3_3():
    client = pymongo.MongoClient("127.0.0.1", 27017)
    mydb = client['school']
    mycol = client['school']['teacher']
    
    query = {"$or":[{"$and":[{"Contact":{"$exists": 0}},{"Role":{"$exists": 0}}]},
                    {"$and":[{'Ratings': {'$lt': 1600}}]}]}
    display = {'_id':1, 'Name':1}
    data = list(mycol.find(query, display))
    client.close()
    
    for row in data:
        #_id,Name,Department,Contact,Role,Ratings
        print("{:<5}{:<20}".format(row["_id"], row["Name"]))

Task_3_3()
```

```python
# Task 3.5 [2]
# Change 1 Teacher from Physics to Computing department

def Task_3_4():
    client = pymongo.MongoClient("127.0.0.1", 27017)
    mydb = client['school']
    mycol = client['school']['teacher']
    
    query = {"Department":"Physics"}
    update_one = {"$set":{"Department":"Computing"}}
    mycol.update_one(query, update_one)
    client.close()

Task_3_4()
```
