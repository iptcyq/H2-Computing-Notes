```python
# Task 1.1

def task1_1(input_value):
    
    # validates input_value, returns -1 if invalid
    
    # checks for valid data types    
    if (type(input_value) != int) and (type(input_value) != str):  
        return -1
    
    # checks for valid length and containing digits 
    input_value = str(input_value)  
    if (len(input_value) != 14) or (not input_value.isdigit()):   
        return -1
      
    # calculate the check digit  
    sum = 0
  
    for i in range(13, -1, -1):   # check every digit in the string, starting from the right
        digit = int(input_value[i])
        if i % 2 == 0:            # digit in even numbered positon
            sum = sum + digit 
        else:                     # digit in odd numbered position
            digit = digit * 2     # double the digit    
            sum = sum + (digit // 10 + digit % 10)   # add the doubled digits to sum
         
    check_digit = sum % 10        # modulo 10 check digit
    if check_digit != 0:
        check_digit = 10 - check_digit
        
    return check_digit

```

```python
# Task 1.2

print (task1_1("50756399655615") == 0)    # string contains valid integer
print (task1_1(12345678901234) == 7)      # valid integer 
print (task1_1(12345) == -1)              # incorrect length
print (task1_1("123456789012ab") == -1)   # string contains non-digit chars
```

```python
# Task 1.3

def task1_3(input_value):
    
    # validates input_value, returns False if invalid
    
    # checks for valid data types    
    if (type(input_value) != int) and (type(input_value) != str):  
        return False
    
    # checks for valid length and containing digits 
    input_value = str(input_value)  
    if (len(input_value) != 15) or (not input_value.isdigit()):   
        return False

    # validate the check digit
    removed_digit = int(input_value[-1])    # remove the last digit   
    check_digit = task1_1(input_value[:14]) # use function task1_1 to calculate the check digit   
                                            # from the remaining 14 digits   
    return (check_digit == removed_digit)
        
    
print (task1_3("507563996556150"))        # string contains valid integer
print (task1_3(123456789012347))          # valid integer 
print (not (task1_3(12345)))              # incorrect length
print (not (task1_3("123456789012ab1")))  # string contains non-digit chars
    
 
```

```python
# Task 2.2

# insertion sort
def task2_2(list_of_integers):  
    
    for i in range(1, len(list_of_integers)):  # iterate from 2nd to last element over the list
        valueToInsert = list_of_integers[i]
        holePos = i
        
        # locate hole position for the element to be inserted
        while (holePos > 0) and (valueToInsert < list_of_integers[holePos-1]):    
            list_of_integers[holePos] = list_of_integers[holePos-1]
            holePos = holePos - 1
            
        # insert the number at hole position
        list_of_integers[holePos] = valueToInsert
    
    return list_of_integers


print (task2_2(task2_1("TEN.txt")))
    
```

```python
# Task 2.3  (ver 2)

# quicksort
def task2_3(list_of_integers):   
    quicksort(list_of_integers, 0, len(list_of_integers)-1)  # call the recursive quick sort on the range[0,len(list_of_integers)-1] 
    return list_of_integers
        
    
def quicksort(int_list, low, high):
    # perform a recursive quicksort
    # int_list is the list of integers passed by reference
    # sorting the range [low,high], inclusive of both ends  
    
    if low < high:   # list has more than one element
 
        # partition into two sublists
        pos = partition(int_list, low, high)  # pos is partitioning index, int_list[pos] is now at right place
 
        # separately sort elements before partition and after partition
        quicksort(int_list, low, pos-1)
        quicksort(int_list, pos+1, high)
     
    # else list has 0 or 1 element and requires no sorting   

    
def partition (int_list, low, high):    
    # split the list into two sublists
    # rearranges the list so that the pivot is properly positioned

    mid = (low + high) // 2   # get the mid index and assign its value to pivot
    pivot = int_list[mid]     
    int_list[mid], int_list[low] = int_list[low], int_list[mid]  # place the pivot in front
                                                    # by exchanging the pivot with the first item
    
    left = low + 1              # index for left search
    right = high                # index for right search

    while left <= right:
        # search from left for element > pivot
        while left <= right and int_list[left] <= pivot:
            left = left + 1

        # search from right for element <= pivot
        while int_list[right] > pivot:
            right = right - 1

        # interchange elements if left and right have not passed each other
        if left < right:
            int_list[left], int_list[right] = int_list[right], int_list[left]


    # end of searches; place pivot in correct position
    int_list[low], int_list[right] = int_list[right], int_list[low]

    return right        # right is the final position of pivot 


print (task2_3(task2_1("TEN.txt")))
```

```python
# Task 2.4

# import the built-in library timeit 
import timeit 

def runTime(file, sort):
    # use the function from Task 2.1 to read in integers from file 
    list_of_integers = task2_1(file)
    # use the timeit function to call the sort function  
    if sort == 'task2_2':
        time = timeit.timeit(lambda: task2_2(list_of_integers), number=1)
    else:
        time = timeit.timeit(lambda: task2_3(list_of_integers), number=1)
    # return the time, in seconds
    return time


def main():
    files = ['TEN.txt', 'HUNDRED.txt', 'THOUSAND.txt']
    print (f"{'':<15}{'INSERTION SORT':<25}{'QUICK SORT'}")
    for file in files:
        insertionSort = runTime(file, 'task2_2')
        quickSort = runTime(file, 'task2_3')
        print (f"{file:<15}{insertionSort:<25}{quickSort}")
      
main()
           
           
           
```

The outputs show that:

* Insertion sort is faster for small data set ('TEN.txt')
* It is much less efficient on large data sets ('HUNDRED.txt'; 'THOUSAND.txt') than quicksort
* For the average time complexity quicksort is O(nlogn) and insertion sort is O(n^2)
* so quicksort is faster than insertion sort - on average.



```python
# Task 3.1

class Node:
    def __init__(self, data, next):
        self.data = data
        self.next = next
        
        
class LinkedList:
    def __init__(self):
        self.head = None
        
        
    def insert(self, integer_value):
        # insert integer_value at the beginning (head) of linked list        
        new_node = Node(integer_value, self.head)
        self.head = new_node
        
        
    def delete(self, integer_value):
        # delete integer_value from the linked list, returns None if integer_value is not present in linked list     

        if not (self.search(integer_value)): # makes use of search() to check if integer_value exists
            return None     
        
        else:
            prev = None
            curr = self.head
            
            while curr.data != integer_value:
                prev = curr
                curr = curr.next
            
            if prev == None:   
                # if integer_value is the first item 
                self.head = curr.next
            else:
                # if integer_value is not the first item 
                prev.next = curr.next     
   
        
    def search(self, integer_value):
        # returns a Boolean value: True if integer_value is in the list, otherwise False   
        probe = self.head
        found = False

        while (probe != None) and (not found):
            if probe.data == integer_value:
                found = True
            else:
                probe = probe.next

        return found
    
            
    def count(self):
        # returns the number of elements in the list, or zero if empty
        probe = self.head
        cnt = 0

        while probe != None:           
            cnt += 1
            probe = probe.next

        return cnt
    
    
    def to_String(self):
        # returns a string containing a suitably formatted list with the elements 
        # separated by a comma and a space, with square brackets at either end

        item_list = []
        probe = self.head
    
        while probe != None:
            item_list.append(str(probe.data))
            probe = probe.next
           
        string = '[' + ", ".join(item_list) + ']'
        return string

    

def main():
    ll = LinkedList()
    
    inputFile = open("Task3data.txt", 'r')
    for line in inputFile:
        ll.insert(int(line.strip()))
    inputFile.close()
        
    print("List:", ll.to_String())
    print("Count:", ll.count())
    print("Search 0:", ll.search(0))
    print("Search 10:", ll.search(10))
    print("Delete 10:", ll.delete(10))
    print ("Delete all:")
    for i in range(10):
        ll.delete(i)
    print("List:", ll.to_String())
    print("Count:", ll.count())

main()
 
```

```python
# Task 3.2

class SortedLinkedList(LinkedList):
    def __init__(self):
        super().__init__() 
        
        
    def insert(self, integer_value):
        # insert integer_value in ascending order into linked list
            
        prev = None
        curr = self.head
        
        while (curr != None) and (integer_value > curr.data):
            prev = curr
            curr = curr.next        
               
        # insert between prev and curr
        new_node = Node(integer_value, curr)      
        
        if prev == None: 
             # if integer_value is the first item 
            self.head = new_node
        else:
            # if integer_value is not the first item 
            prev.next = new_node
            

def main():
    ll = SortedLinkedList()
    
    inputFile = open("Task3data.txt", 'r')
    for line in inputFile:
        ll.insert(int(line.strip()))
    inputFile.close()
        
    print("List:", ll.to_String())
    print("Search 94:", ll.search(94))
    
    
main()
```

```python
# Task 3.3 (ver 1: 1st linked element is the last element in the queue)

class Queue(LinkedList):     
    def __init__(self):
        super().__init__()     
        
        
    def enqueue(self, integer_value):
        # insert integer_value to the end of the queue
        self.insert(integer_value)
        
        
    def dequeue(self):
        # return the first element in the queue. If the queue is empty, return None
        if self.count() == 0:
            return None 
        else:
            probe = self.head               
            for i in range(self.count()-1):
                probe = probe.next
        
            self.delete(probe.data)
            return probe.data   
        

def main():
    q = Queue()
    
    inputFile = open("Task3data.txt", 'r')
    for line in inputFile:
        q.enqueue(int(line.strip()))
    inputFile.close()
  
    print("Queue (rear to front):", q.to_String())
    print ("First five elements to be dequeued:", end = ' ')        
    for i in range(5):
        print (q.dequeue(), end = '  ')
    
        
main()

```

```python
# Task 3.3 (ver 2: 1st linked element is the first element in the queue)

class Queue(LinkedList):     
    def __init__(self):
        super().__init__()
     
        
    def enqueue(self, integer_value):
        # insert integer_value to the end of the queue
        new_node = Node(integer_value, None)        
            
        if self.count() == 0:
            self.head = new_node
        else:
            probe = self.head
            for i in range(self.count()-1):
                probe = probe.next
            probe.next = new_node
            
    
    def dequeue(self):
        # return the first element in the queue. If the queue is empty, return None
        if self.count() == 0:
            return None
        else:
            removed = self.head.data                                    
            self.delete(removed)  # self.head = self.head.next 
            return removed   
         

def main():
    q = Queue()
    
    inputFile = open("Task3data.txt", 'r')
    for line in inputFile:
        q.enqueue(int(line.strip()))
    inputFile.close()

    print("Queue (front to rear):", q.to_String())   
    print ("First five elements to be dequeued:", end = ' ')        
    for i in range(5):
        print (q.dequeue(), end = '  ')

main()

```
