```python
# Task 4.1
import sqlite3

# create database if not exist
conn = sqlite3.connect('LIBRARY.db')

""" For debugging """
conn.execute("DROP TABLE IF EXISTS Book")
conn.execute("DROP TABLE IF EXISTS Member")
conn.execute("DROP TABLE IF EXISTS Loan")
# not required, but table cannot be created if it exists, so it is better to DROP TABLE before creating

create_book = """CREATE TABLE Book (
    BookID INTEGER UNIQUE,
    Title TEXT,
    Genre TEXT,
    PRIMARY KEY(BookID)
);"""

create_member = """CREATE TABLE Member (
    MemberNumber INTEGER UNIQUE,
    FamilyName TEXT,
    GivenName TEXT,
    PRIMARY KEY(MemberNumber)
);"""

create_loan = """CREATE TABLE Loan (
    LoanID INTEGER UNIQUE,
    MemberNumber INTEGER,
    BookID INTEGER,
    DateLoaned TEXT,
    Returned TEXT,
    PRIMARY KEY(LoanID),
    FOREIGN KEY(MemberNumber) REFERENCES Member(MemberNumber),
    FOREIGN KEY(BookID) REFERENCES Book(BookID)
);"""

conn.execute(create_book)
conn.execute(create_member)
conn.execute(create_loan)

conn.commit()
conn.close()
```

```python
# Task 4.2

conn = sqlite3.connect('LIBRARY.db')

# read from BOOK.txt
bookfile = open('BOOK.txt', 'r')
bookdata = bookfile.read().strip().split('\n')
bookfile.close()

for line in bookdata:
    bookid, title, genre = line.strip().split(',')
    
    # insert into Book table in db
    conn.execute('INSERT INTO Book (BookID, Title, Genre) ' + \
                'VALUES (?, ?, ?)', (bookid, title, genre))

# read from MEMBER.txt
memberfile = open('MEMBER.txt', 'r')
memberdata = memberfile.read().strip().split('\n')
memberfile.close()

for line in memberdata:
    memberID, family, given = line.strip().split(',')
    
    # insert into Member table in db
    conn.execute('INSERT INTO Member (MemberNumber, FamilyName, GivenName) ' + \
                'VALUES (?, ?, ?)', (memberID, family, given))

# read data from LOAN.txt
loanfile = open('LOAN.txt', 'r')
loandata = loanfile.read().strip().split('\n')
loanfile.close()

for line in loandata:
    loanID, memberID, bookID, loan, returned = line.strip().split(',')
    
    # insert into db
    conn.execute('INSERT INTO Loan (LoanID, MemberNumber, BookID, DateLoaned, Returned) ' + \
                'VALUES (?, ?, ?, ?, ?)', (loanID, memberID, bookID, loan, returned))

conn.commit()    # commit all edits at the end
conn.close()
```

```python
# task 4.3

conn = sqlite3.connect('LIBRARY.db')
conn.row_factory = sqlite3.Row

memberID = int(input('Member number: '))

# execute search query
cursor = conn.execute('SELECT Book.Title, Loan.Returned ' +
        'FROM Book INNER JOIN Loan ' +
        'ON Book.BookID = Loan.BookID ' +
        'WHERE Loan.MemberNumber = ?', (memberID,))

# # print out results
print(f"\n{'Title':<30}Returned")
for result in cursor:
    print(f"{result['Title']:<30}{result['Returned']}")

conn.close()

# use member number 200
```
