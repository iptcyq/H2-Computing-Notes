```python
# Task 3.1

class Tree():
    def __init__(self,data):
        self.data = data
        self.left_pointer = None
        self.right_pointer = None
    # 4 marks until here

    # 6 marks for store()
    def store(self, new):
        if self.data == None: # store into empty tree
            self.data = new
        elif self.data > new: # if new is smaller
            if self.left_pointer == None: # if nothing on left
                self.left_pointer = Tree(new) # put into left as new tree
            else: # if left has a subtree
                self.left_pointer.store(new) # recursively store in left subtree
        elif self.data < new:
            if self.right_pointer == None:
                self.right_pointer = Tree(new)
            else:
                self.right_pointer.store(new)
                
    # 7 marks for both in_order() and post_order() and calling later
    def in_order(self):
        if self.data != None:
            if self.left_pointer != None:
                self.left_pointer.in_order()
            print(self.data, end = ' ')
            if self.right_pointer != None:
                self.right_pointer.in_order()
                
    def post_order(self):
        if self.data != None:
            if self.left_pointer != None:
                self.left_pointer.post_order()
            if self.right_pointer != None:
                self.right_pointer.post_order()
            print(self.data, end = ' ')

            
# MAIN (5 marks)
from random import randint
new_tree = Tree(None)
for i in range(10):
    rand = randint(0,999)
    new_tree.store(rand)

# MAIN (2 marks)
print("In-order: ")
new_tree.in_order()
print("\nPost-order: ")
new_tree.post_order()
```

```python
# Task 3.1 (alternative, where subtree is fed into the tree.methods)

class Tree2():
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    # 4 marks until here

    # 6 marks for store()
    def store(self, new, subtree):
        if self.data == None: # store into empty tree
            self.data = new
        elif subtree.data > new: # if new is smaller
            if subtree.left == None: # if nothing on left
                subtree.left = Tree2(new) # put into left as new tree
            else: # if left has a subtree
                self.store(new, subtree.left) # recursively store in left subtree
        elif subtree.data < new:
            if subtree.right == None:
                subtree.right = Tree2(new)
            else:
                self.store(new, subtree.right)
                
    # 7 marks for both in_order() and post_order() and calling later
    def in_order(self, subtree):
        if subtree != None:
            self.in_order(subtree.left)
            print(subtree.data, end = ' ')
            self.in_order(subtree.right)
    def post_order(self, subtree):
        if subtree != None:
            self.post_order(subtree.left)
            self.post_order(subtree.right)
            print(subtree.data, end = ' ')

# MAIN (5 marks)
from random import randint
new_tree = Tree2(None)
for i in range(10):
    rand = randint(0,999)
    new_tree.store(rand, new_tree)

# MAIN (2 marks)
print("In-order: ")
new_tree.in_order(new_tree)
print("\nPost-order: ")
new_tree.post_order(new_tree)
```
