```python
# Task 4.1
import random

def init():
    board = [[ '.' , '.' , '.' , '.' ],
             [ '.' , '.' , '.' , '.' ],
             [ '.' , '.' , '.' , '.' ],
             [ '.' , '.' , '.' , '.' ]]
    x = random.randint(1,4)
    y = random.randint(1,4)
    board[x-1][y-1] = 'T'
    return board
    
# init()
```

```python
# Task 4.2
def displayBoard(board):
    for row in board:
        for cell in row:
            print(cell, end=' ')
        print()

# displayBoard(board)
```

```python
# Task 4.3
def getPlayerMove(board, mark):
    valid = False
    while not valid:
        # check x and y are valid numbers
        x = input("Enter x coordinate: ")
        y = input("Enter y coordinate: ")
        if x.isnumeric() and y.isnumeric(): 
            # forgot to check IN RANGE
            if 1 <= int(x) <= 4 or 1 <= int(y) <= 4:
                valid = True
            # check if x,y are occupied
            if board[int(y)-1][int(x)-1] == '.': 
                valid = True
            else:
                valid = False
                # should print error: cell taken

    board[int(y)-1][int(x)-1] = mark

# getPlayerMove(board,'X')
# displayBoard(board)
```

```python
# Task 4.4
def checkWin(board):
    # check row
    for x in range(4):
        if '.' not in board[x]: # full row
            if 'O' not in board[x] or 'X' not in board[x]: # only O or X
                return True
    
    # check cols
    for y in range(4):
        col = []
        for i in range(4):
            col.append(board[i][y])
        if '.' not in col: # full row
            if 'O' not in col or 'X' not in col: # only O or X
                return True

    # check left diagonal
    diag = []
    for d in range(4):
        diag.append(board[d][d])
    if '.' not in diag: # full row
            if 'O' not in diag or 'X' not in diag: # only O or X
                return True
    
    # right diagonal
    diag = []
    for d in range(4):
        diag.append(board[d][3-d])
    if '.' not in diag: # full row
            if 'O' not in diag or 'X' not in diag: # only O or X
                return True
    return False

# checkWin(board)
```

```python
def main():
    board = init()
    displayBoard(board)
    turn = 0
    
    end = False
    while not end:
        turn += 1
        if turn % 2 == 1: # X turn
            print("Player X's turn: ")
            getPlayerMove(board,'X')
        else: # Y turn
            print("Player O's turn: ")
            getPlayerMove(board,'O')
        displayBoard(board)
        
        end = checkWin(board) # end?
        if not end: # draw?
            end = True
            for row in board:
                if '.' in row:
                    end = False
            if end:
                turn = 0

    if turn == 0:
        print("Draw.")
    elif turn % 2 == 1: # X won
        print("Player X won")
    else:
        print("Player O won")
```
