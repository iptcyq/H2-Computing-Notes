```python
# Task 2.1
import datetime

class LibraryItem():
    def __init__(self, item_id, title, loaned_by="", due_date=""):
        self.item_id = item_id
        self.title = title
        self.loaned_by = loaned_by
        self.due_date = due_date
        
    def is_on_loan(self):
        return self.loaned_by != ""
        
    def return_item(self):
        self.loaned_by = ""
        
    def loan_to(self, user_id):
        if not self.is_on_loan():
            self.loaned_by = user_id
            today = datetime.date.today()
            newdate = today + datetime.timedelta(weeks=3)
            self.due_date = newdate.strftime('%d-%m-%Y')
            return True
        else: # is already loaned
            return False
    
    def print_details(self):
        print("Title:",self.title)
        print("Loaned by:",self.loaned_by)
        print("Due date:",self.due_date)
```

```python
# Task 2.2
class Book(LibraryItem):
    def __init__(self, item_id, title, author, category, loaned_by="", due_date="", reserved_by=""):
        super().__init__(item_id, title, loaned_by, due_date)
        self.author = author
        self.category = category
        self.reserved_by = reserved_by
        
    def is_reserved(self):
        return self.reserved_by != ""
    
    def release(self):
        self.reserved_by = ""
        
    def reserve_for(self, user_id):
        if not self.is_reserved():
            self.reserved_by = user_id
            return True
        else:
            return False
    
    def print_details(self):
        print("Title:",self.title)
        print("Loaned by:",self.loaned_by) if self.is_on_loan() else print("Loaned by: -")
        print("Due date:",self.due_date) if self.is_on_loan() else print("Due date: -")
        print("Author:",self.author)
        print("Category:",self.category)
        print("Reserved for:",self.reserved_by) if self.is_reserved() else print("Reserved for: -")
        
class CompactDisc(LibraryItem):
    def __init__(self, item_id, title, artist, genre, loaned_by="", due_date="",):
        super().__init__(item_id, title, loaned_by, due_date)
        self.artist = artist
        self.genre = genre
        
    def print_details(self):
        print("Title:",self.title)
        print("Loaned by:",self.loaned_by) if self.is_on_loan() else print("Loaned by: -")
        print("Due date:",self.due_date) if self.is_on_loan() else print("Due date: -")
        print("Artist:",self.artist)
        print("Genre:",self.genre)
```

```python
# Task 2.3
f = open("../resource/libraryitems.txt")
data = f.read().strip().split('\n')
f.close()

libraryitems = []

for item in data:
    item_id, title, first, second = item.split(',')
    if item_id[0] == 'B': # book
        new = Book(item_id, title, first, second)
    else: # cd
        new = CompactDisc(item_id, title, first, second)
    new.print_details()
    print()
    libraryitems.append(new)
```

```python
for item in libraryitems:
    item.loan_to("S1111111H")
    if type(item) == Book:
        item.reserve_for("S1222222D")
        
    item.print_details()
    print()
```
