```python
# Task 1.3
def task1_3(input_value):
    code = input_value
    all_values = []

    for i in range(4):
        val = ord(code[i]) + ord(code[i])//11 - 60
        all_values.append(val)

    for j in range(4,10):
        val = int(code[j])
        all_values.append(val)

    sum = 0
    for k in range(10):
        sum += all_values[k] * (2**k)

    return sum%11
```

```python
# Task 1.4
def task1_4(input_value):
    if task1_1(input_value[:-1]) == True: # valid
        check_dig = task1_3(input_value[:-1])
        if check_dig == int(input_value[-1]):
            return True
    return False

print(task1_4('PONU2079674'))
print(task1_4('XONU2079674'))
```

```python
# Task 2.2
f = open("SPORTS-FACILITIES.txt")
data = f.read().strip().split('\n')
f.close()

sortd = []
for line in data:
    fac, bookings = line.split(',')
    sortd.append((fac, int(bookings)))
    
# quick sort algo
def quicksort(arr, low, high):
    if len(arr) <= 1 or low >= high:
        return arr
    else:
        sep = low+1 # separator
        for i in range(low+1, high):
            if arr[i][1] > arr[low][1]:
                arr[i], arr[sep] = arr[sep], arr[i]
                sep += 1
                
        arr[sep-1], arr[low] = arr[low], arr[sep-1]

        quicksort(arr, low, sep-1)
        quicksort(arr, sep, high)

quicksort(sortd, 0, len(sortd))
print(f"Facility Type \t\tNumber of bookings")
print("------------------------------------------")
for item in sortd:
    print(f"{item[0]} \t\t{item[1]}")
```

```python
# Task 3.1
class Process():
    def __init__(self, name, remaining):
        self.name = name
        self.remaining = remaining
        
    def execute(self, t):
        self.remaining -= t
        
    def is_completed(self):
        return self.remaining <= 0
```

```python
# Task 3.2
class CircularQueue():
    def __init__(self, size):
        self.items = [None] * size
        self.front = -1
        self.rear = -1
        
    def is_empty(self):
        for item in self.items:
            if item != None:
                return False
        return True
    
    def is_full(self):
        for item in self.items:
            if item == None:
                return False
        return True
        
    def enqueue(self, process):
        if self.is_full():
            print("Error: queue is full")
        else:
            self.rear = (self.rear+1)%len(self.items)
            self.items[self.rear] = process
            if self.front == -1:
                self.front = 0
             
    def dequeue(self):
        if self.is_empty():
            print("Error: queue is empty")
        else:
            proc = self.items[self.front]
            self.items[self.front] = None
            self.front = (self.front+1)%len(self.items)
            if self.is_empty():
                self.front = self.rear = -1
            return proc
                
    def display(self):
        print("Displaying Queue:")
        if self.is_empty():
            print("Queue is empty")
        if self.rear < self.front:
            for i in range(self.front, len(self.items)):
                print(self.items[i], end=",")
            for j in range(0, self.rear+1):
                print(self.items[j], end=",")
        else:
            for k in range(self.front, self.rear+1):
                print(self.items[k], end=",")
        print()
```

```python
# Task 3.3
class FCFSScheduler(CircularQueue):
    def start(self):
        end = False
        
        while not end:
            proc = self.dequeue()
            while not proc.is_completed():
                proc.execute(0.5)
                print(proc.name, "\b:", proc.remaining, "seconds remaining")
            print(proc.name, "completed")
            if self.is_empty():
                end = True
                
class RRScheduler(CircularQueue):
    def start(self):
        end = False
        
        while not end:
            proc = self.dequeue()
            proc.execute(0.5)
            if proc.is_completed():
                print(proc.name, "completed")
            else:
                print(proc.name, "\b:", proc.remaining, "seconds remaining")
                self.enqueue(proc)
            if self.is_empty():
                end = True
```

```python
# Task 3.4
f = open("CPU-PROCESSES.csv")
data = f.read().strip().split('\n')
f.close()

fcfs = FCFSScheduler(len(data)-1)
for proc in data[1:]:
    name, time = proc.split(',')
    newproc = Process(name, float(time))
    fcfs.enqueue(newproc)
    
print("Starting FCFS Scheduler:")
fcfs.start()
```
