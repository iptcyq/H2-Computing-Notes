```python
# Task 3
class Node:
    def __init__(self, data, ptr):
        self.data = data
        self.ptr = ptr

class DataStructure:
    def __init__(self):
        self.Head = None
        self.Tail = None
        
    def display(self):
        if self.Head == None:
            print("Empty Linked List")
        else:
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
        print()
        
class Stack(DataStructure):
    def insert(self,value):
        if self.Head == None:
            self.Tail = Node(value, self.Head)
        self.Head = Node(value, self.Head)
        
        
    def delete(self):
        if self.Head == None:
            print('Stack is empty')
        else:
            data = self.Head.data
            self.Head = self.Head.ptr
            if self.Head == None:
                self.Tail = None
            return data
        
    def display(self):
        if self.Head == None:
            print("Empty Stack!")
        else:
            print("Stack Contents:")
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
            
        print()

class Queue(DataStructure):
    def insert(self, value):
        if self.Head == None: # insert into empty queue
            self.Head = Node(value,self.Head)
            self.Tail = self.Head
        else:
            self.Tail.ptr = Node(value, None)
            self.Tail = self.Tail.ptr
        
    def delete(self):
        if self.Head == None:
            print('Queue is empty')
        else:
            data = self.Head.data
            self.Head = self.Head.ptr
            if self.Head == None:
                self.Tail = None
            return data
    
    def display(self):
        if self.Head == None:
            print("Empty Queue!")
            print(self.Tail.data)
        else:
            print("Queue Contents:")
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
        print()

s = Stack()
q = Queue()
with open('../Additional_Materials/TASK3stack.txt','r') as f:
    for line in f:
        line = line.strip()
        s.insert(line)
with open('../Additional_Materials/TASK3queue.txt','r') as f:
    for line in f:
        line = line.strip()
        q.insert(line)
s.display()
q.display()

s.delete()
s.delete()
q.delete()
q.delete()

print("After deleting twice:")
s.display()
q.display()
```
