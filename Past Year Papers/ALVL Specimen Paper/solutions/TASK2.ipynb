```python
# Task 2.1

# finding position to split
def split(name_list, low, high):
    pivot = low
    left = low
    right = high
    
    while left < right: 
        while left < right and name_list[left] < name_list[pivot]:
            left += 1
    
        while right > left and name_list[right] > name_list[pivot]:
            right -= 1
    
        if left < right:
            temp = name_list[left]
            name_list[left] = name_list[right]
            name_list[right] = temp
     
    name_list[pivot], name_list[right] = name_list[right], name_list[pivot]
    
    return right
    
# recurisve quicksort
def quicksort(name_list, low, high):
    if low < high:
        pos = split(name_list, low, high)
        quicksort(name_list, low, pos - 1)
        quicksort(name_list, pos + 1, high)
    
    
# generate list of filenames
with open("TESTDATA.txt") as f:
    name_list = []
    content_dict = {}
    for line in f:
        name, date, size = line.strip().split(',')
        name_list.append(name)
        
quicksort(name_list, 0, len(name_list) - 1)

print("List of filenames:")
for name in name_list:
    print(name)
    
with open("TESTDATA_Filename_order.txt",'w') as f:
    for name in name_list:
        f.write(name+'\n')
        
        
```

```python
# Task 2.2

# bubble sort
def bubble(date_list):
    for i in range(len(date_list)):
        for j in range(len(date_list) - 1):
            if date_list[j] > date_list[j + 1]:
                temp = date_list[j]
                date_list[j] = date_list[j + 1]
                date_list[j + 1] = temp


def sort_display():
    # generate list of dates, dictionary with date as key 
    # and [name, size] as value
    with open("TESTDATA.txt") as f:
        date_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            
            if date not in date_list: # some dates are repeated
                date_list.append(date)
                
            if date in content_dict:
                content_dict[date].append([name, size])
            else:
                content_dict[date] = [[name, size]]
    
    bubble(date_list)
    
    print('{:<20} {:<20} {}'.format('Date Modified', 'File Name', 'Size in KB'))
    
    for date in date_list:
        for name, size in content_dict[date]:
            print('{:<20} {:<20} {}'.format(date, name, size))
              

# search with given name
def search_name():
    name_given = input("Enter filename:")
    with open("TESTDATA.txt") as f:
        name_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            name_list.append(name)
            content_dict[name] = size
    
    if name_given in name_list:
        print("File name:", name_given)
        print("File size:", content_dict[name_given], 'KB')
    else:
        print("File {} not found.".format(name_given))



# search with given date
def search_date():
    date_given = input("Enter modified date:")
    with open("TESTDATA.txt") as f:
        date_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            if date not in date_list:
                date_list.append(date)
            if date in content_dict:
                content_dict[date].append(name)
            else:
                content_dict[date] = [name]
            
    if date_given in date_list:
        print("File(s) modified on this date:", date_given)
        for name in content_dict[date_given]:
            print(name)
    else:
        print("No files modified on", date_given)


print('Test case 1: sort in date order')
sort_display()
print()

print('Test case 2: search for a file name that is contained in the file')
search_name()
print()

print('Test case 3: search for a file name that is not contained in the file')
search_name()
print()

print("Test case 4: display files modified on a given date")
search_date()
print()

print("Test case 5: display message showing that no files were modified on a given date")
search_date()
```
