```python
# Task 4.4: different URL for get and post methods

import flask
from flask import render_template, request
import sqlite3

app = flask.Flask(__name__)

@app.route('/')
def home():
    return render_template('home2.html')

@app.route('/filter', methods = ['POST'])
def filter():
    db = sqlite3.connect('equipment.db')
    location = request.form['location']
    query = "SELECT SerialNumber, Type FROM Device WHERE Location = ? AND NOT WrittenOff"
    results = db.execute(query, (location,)).fetchall()
    db.close()
    
    return render_template('filter.html', results = results)
    
if __name__ == '__main__':
    app.run()
```

```python
# Task 4.4: same URL for get and post methods

import flask
from flask import render_template, request
import sqlite3

app = flask.Flask(__name__)

@app.route('/', methods = ['GET', 'POST'])
def combine():
    if request.method == 'GET':
        return render_template('home1.html')
    if 'location' in request.form:
        db = sqlite3.connect('equipment.db')
        location = request.form['location']
        query = "SELECT SerialNumber, Type FROM Device WHERE Location = ? AND NOT WrittenOff"
        results = db.execute(query, (location,)).fetchall()
        db.close()

        return render_template('filter.html', results = results)
    
if __name__ == '__main__':
    app.run()
```
