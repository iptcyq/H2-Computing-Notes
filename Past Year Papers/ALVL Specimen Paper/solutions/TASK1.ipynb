```python
# Task 1.1

letter = input("Enter a letter:")

# one single letter
cond = letter.isalpha() and len(letter) == 1

while not cond:
    print("Please enter a single letter in the range A-Z or a-z.")
    letter = input("Enter a letter:")
    cond = letter.isalpha() and len(letter) == 1

print("Letter:", letter)
   
```

```python
# Task 1.2


letter = input("Enter a letter:")

base = int(input("Enter a number base:"))
# ensure base in correct range
while base <= 10 or base >= 15:
    print("Please enter a base between 11 and 14 (inclusive).")
    base = input("Enter a number base:")

# match the letters for base 
base_match = {10:'A', 11:'B', 12:'C', 13:'D'}
for i in range(10):
    base_match[i] = str(i)

# perform convertion
value = ord(letter)
base_value = ''
while value != 0:
    new = base_match[value % base]
    base_value = new + base_value
    value = value // base

# display results
print('{:<20} {}'.format('Letter', letter))
print('{:<20} {}'.format('Denary', ord(letter)))
print('{:<20} {}'.format('Number Base ' + str(base), base_value))
```

```python
# Task 1.3

def display():
    print("1: Enter a letter")
    print("2. Convert to Denary")
    print("3. Convert to Base 11")
    print("4. Convert to Base 12")
    print("5. Convert to Base 13")
    print("6. Convert to Base 14")
    print("7. End")
    
def convert(letter, base):
    value = ord(letter)
    base_match = {10:'A', 11:'B', 12:'C', 13:'D'}
    for i in range(10):
        base_match[i] = str(i)
        
    base_value = ''
    while value != 0:
        new = base_match[value % base]
        base_value = new + base_value
        value = value // base
    return base_value   
        
display()
option = int(input("Enter your option (1-7):"))

while option != 7:
    if option == 1:
        letter = input("Letter:")
        print()
    elif option == 2:
        print("Denary:", ord(letter))
        print()
    elif option == 3:
        print("Number Base 11:", convert(letter, 11))
        print()
    elif option == 4:
        print("Number Base 12:", convert(letter, 12))
        print()
    elif option == 5:
        print("Number Base 13:", convert(letter, 13))
        print()
    elif option == 6:
        print("Number Base 14:", convert(letter, 14))
        print()
    else:
        print("Error! Please enter option 1 - 7. ")
    option = int(input("Enter your option:"))
```
