```python
# Task 4.1
class Node:
    def __init__(self, data, idx, next=None):
        self.data = data
        self.index = idx
        self.next = next
    def set_data(self, data):
        self.data = data
    def get_data(self):
        return self.data
    def set_next(self, next):
        self.next = next
    def get_next(self):
        return self.next

class LinkedStructure:
    def __init__(self):
        self.Initialise()
    def Initialise(self):
        self.Head = None
        self.Tail = None
        
        self.Free = Node("", 1)
        self.Array = [self.Free]
        current = self.Free
        for i in range(2, 12):
            new = Node("", i)
            self.Array.append(new)
            current.set_next(new)
            current = new
    
    def Add(self, item):
        if self.Free == None: # no free space
            print("Error: Array full")
        else:
            # take a node from free space
            new = self.Free
            new.set_data(item)
            self.Free = new.get_next()
            new.set_next(None)
            
            if self.Head == None: # no items
                new.set_data(item)
                self.Tail = self.Head = new
            elif item < self.Head.get_data(): # first item
                new.set_next(self.Head)
                self.Head = new
            else:
                # find the correct position for the new node
                current = self.Head
                while current.get_next() != None and current.get_next().get_data() < item:
                    # go through the list
                    current = current.get_next()
                if current == None: # last item
                    self.Tail.set_next(new)
                    self.Tail = new
                else:
                    # insert the new node here
                    new.set_next(current.get_next())
                    current.set_next(new)

    def Remove(self, item): # NOT DONE
        if self.Head == None: # no items
            print("Error: Array empty")
            return
        elif self.Head.get_data() == item: #item at head
            freeNode = self.Head
            self.Head = freeNode.get_next()
        else: # more than 1 
            curr = self.Head
            while curr != None and curr.get_next().get_data() != item:
                curr = curr.get_next()
            if curr == None:
                print("Error: Item not found")
                return
            freeNode = curr.get_next()
            curr.set_next(freeNode.get_next())
        
        # return node to free space list
        freeNode.set_next(self.Free)
        self.Free = freeNode
    
    def Display(self):
        current = self.Head
        while current != None:
            print(current.get_data())
            current = current.get_next()
    def PrintStructure(self): # NOT DONE
        # i might need to restart this cos its wrong
        # I'll just do what might be the answer if i wasnt misinterpreting it
        curr_idx = 1
        current = self.Head
        while current != None: # NOT DONE need to fix when idx 1 is removed
            if current.index == curr_idx:
                print(f"{current.index}: {current.get_data()}", end=" ")
                print(f"-> {current.get_next().index}") if current.get_next() != None else print(f"-> 0")
                curr_idx += 1
                current = self.Head
            else:
                current = current.get_next()
    
    def IsEmpty(self):
        return self.Head == None
    def IsFull(self):
        return self.Free == None
```

```python
# Redoing the entire task
# Task 4.1
class Node():
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
        
class LinkedStructure():
    def __init__(self):
        self.arr = [Node("",2),Node("",3),Node("",4),Node("",5),Node("",6),Node("",7),Node("",8),Node("",9),Node("",10),Node("",0)]
        self.Head = self.Tail = -1
        self.Free = 1
    
    def Add(self, item):
        if self.Free == -1: # list is full
            print("Error: List is full")
        elif self.Tail == -1: # list is empty
            self.Head = self.Tail = self.Free
            self.Free = self.arr[self.Tail-1].next
            self.arr[self.Tail-1].data = item
            self.arr[self.Tail-1].next = 0
        elif self.arr[self.Head-1].data > item: # first item
            new_node = self.Free
            self.Free = self.arr[new_node-1].next
            self.arr[new_node-1].data = item
            self.arr[new_node-1].next = self.Head
            self.Head = new_node
        else:
            curr = self.Head
            while self.arr[curr-1].next != 0 and self.arr[self.arr[curr-1].next-1].data < item:
                curr = self.arr[curr-1].next
            if self.arr[curr-1].next == 0:
                self.Tail = self.Free
            next_free = self.arr[self.Free-1].next # set next free
            self.arr[self.Free-1].next = self.arr[curr-1].next # reset curr next
            self.arr[curr-1].next = self.Free # reset prev next to curr
            self.arr[self.Free-1].data = item # change data
            self.Free = next_free # change free
    
    def Remove(self, item):
        if self.Tail == -1: # empty list
            print("Error: List is full")
        elif self.Head == self.Tail: # one item left
            if self.arr[self.Head-1] == item:
                self.arr[self.Head-1].data = ""
                self.arr[self.Head-1].next = self.Free
                self.Free = self.Head
                self.Head = self.Tail = -1
            else:
                print("Error: Item not in list")
        elif self.arr[self.Head-1].data == item:
            new_free = self.Head
            self.Head = self.arr[self.Head-1].next
            self.arr[new_free-1].data = ""
            self.arr[new_free-1].next = self.Free
            self.Free = new_free
        else: 
            curr = self.Head
            while self.arr[curr-1].next != 0 and self.arr[self.arr[curr-1].next-1].data < item:
                curr = self.arr[curr-1].next
            if self.arr[self.arr[curr-1].next-1].data == item:
                new_free = self.arr[curr-1].next
                self.arr[curr-1].next = self.arr[new_free-1].next
                self.arr[new_free-1].data = ""
                self.arr[new_free-1].next = self.Free
                self.Free = new_free
            else:
                print("Error: Item not in list")
    
    def Display(self):
        curr = self.Head
        while curr != 0:
            print(self.arr[curr-1].data)
            curr = self.arr[curr-1].next
    def PrintStructure(self):
        print("Head:",self.Head)
        print("Tail:",self.Tail)
        print("Free:",self.Free)
        
        count = 1
        for i in self.arr:
            print(f"{count}: {i.data} -> {i.next}")
            count += 1
    
    def IsEmpty(self):
        return self.Head == -1
    def IsFull(self):
        return self.Free == -1
```

```python
# Task 4.3
class Queue(LinkedStructure):
    def __init__(self):
        super().__init__()
    
    def Add(self, item):
        if self.Free == -1: # full list
            print("Error: Queue full")
        elif self.Head == -1: # empty list
            self.Head = self.Tail = self.Free
            self.Free = self.arr[self.Head-1].next
            self.arr[self.Head-1].data = item
        else:
            new = self.Free
            self.Free = self.arr[self.Free-1].next
            self.arr[self.Tail-1].next = new
            self.arr[new-1].next = 0
            self.arr[new-1].data = item
            self.Tail = new
    def Remove(self):
        if self.Head == -1: # empty list
            print("Error: Empty list")
            return
        elif self.Head == self.Tail: # 1 item
            self.arr[self.Head-1].next = self.Free
            item = self.arr[self.Head-1].data
            self.arr[self.Head-1].data = ""
            self.Free = self.Head
            self.Head = self.Tail = -1
        else:
            new_free = self.Head
            self.Head = self.arr[new_free-1].next
            self.arr[new_free-1].next = self.Free
            self.Free = new_free
            item = self.arr[new_free-1].data
            self.arr[new_free-1].data = ""
        return item
            
    def Display(self):
        curr = self.Head
        while curr != 0:
            print(self.arr[curr-1].data)
            curr = self.arr[curr-1].next
```
