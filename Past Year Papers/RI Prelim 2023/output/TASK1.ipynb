```python
# Task 1.1
def task1_1(filename):
    f = open(filename)
    data = f.read().split('\n\n')
    f.close()
    
    all_ingredients = {}
    for recipe in data:
        for ingredient in recipe.split('\n')[1:]:
            name, qty, unit = ingredient.split(',')
            if name not in all_ingredients.keys():
                all_ingredients[name] = 1
            else:
                all_ingredients[name] += 1
    
    highest_val = 0
    highest_val_ingredient = ''
    for ingredient in all_ingredients.keys():
        if all_ingredients[ingredient] > highest_val:
            highest_val = all_ingredients[ingredient]
            highest_val_ingredient = ingredient
    
    return highest_val_ingredient, highest_val

print(task1_1('RECIPES.TXT'))
```

```python
# Task 1.2
def task1_2(strings):
    for i in range(len(strings)):
        curr = i
        while curr > 0 and strings[curr] < strings[curr-1]:
            strings[curr], strings[curr-1] = strings[curr-1], strings[curr]
            curr -= 1
    return strings
```

```python
# Task 1.3
def task1_3(filename):
    f = open(filename)
    data = f.read().split('\n\n')
    f.close()
    
    all_ingredients = {}
    for recipe in data:
        for ingredient in recipe.split('\n')[1:]:
            name, qty, unit = ingredient.split(',')
            if name not in all_ingredients.keys():
                all_ingredients[name] = {unit:float(qty)}
            elif unit not in all_ingredients[name].keys():
                all_ingredients[name] = {unit:float(qty)}
            else:
                all_ingredients[name][unit] += float(qty)
            
    all_ingredient_names = task1_2(list(all_ingredients.keys()))
    for item in all_ingredient_names:
        print(f"{item} has")
        for qty_type in all_ingredients[item]:
            print(f"{all_ingredients[item][qty_type]} {qty_type}")
        print()
        
task1_3('RECIPES.TXT')
```
