# Sorts and searches
Note: A levels will complicate things with a bundle of data, and it must be sorted wrt a certain index.

Example: Given a bunch of data in the format `[<name>, <age>, <height>]`, sort in ascending order of age.

```python
# bubble sort, O(n) = n^2
def bubble(alist):
    length = len(alist)
    for i in range(length): # outer loop, runs n times
        for j in range(length - i - 1): # inner loop, runs 1 less time each round
            if alist[j] > alist[j+1]:
                alist[j], alist[j+1] = alist[j+1], alist[j]
    return alist

print(bubble([1,4,5,3,0]))

# possible improvements: putting a flag to check if swaps were made in latest run
# abs  
```

```python
# merge sort, O(n) = n * log(n)

def merge(alist):
    if len(alist) == 1:
        return alist
    length = len(alist)
    left = alist[:length//2]
    right = alist[length//2:]
    left = merge(left)
    right = merge(right)
    ret = []
    
    while len(left) != 0 and len(right) != 0:
        if left[0] < right[0]:
            ret.append(left[0])
            left = left[1:]
        else:
            ret.append(right[0])
            right = right[1:]
    
    while len(left) != 0:
        ret.append(left[0])
        left = left[1:]
    
    while len(right) != 0:
        ret.append(right[0])
        right = right[1:]
        
    return ret
```

```python
# quick sort, O(n) = n^2

def dirty_quicksort(alist): # non-mutator, returns
    if len(alist) <= 1:
        return alist
    left = []
    right = []
    pivot = alist[-1]
    alist = alist[:-1]
    for i in alist:
        if i < pivot:
            left.append(i)
        else:
            right.append(i)
    return dirty_quicksort(left) + [pivot] + dirty_quicksort(right)

```

```python
# insertion sort
def insertion_sort(array):
  for i in range(1,len(array)):
    temp = array[i] # store in temporary location (trying to insert this thing)
    pointer = i - 1 # start looking in position on the left

    while pointer >= 0 and temp < array[pointer]: # put edge case on the left of "and"
      array[pointer + 1] = array[pointer] # shifting element at pointer to the right
      pointer -= 1 # decrement pointer, move to left by 1

    array[pointer+1] = temp # insert here (on the right of first smaller element)

  return array

print(insertion_sort([5,1,99,23,3,2,0]))
```

```python
class Node:
    def __init__(self, data, next = None):
        self.data = data
        self.next = next
```

```python
class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
        
    def enq(self, value):
        if self.size == 0:
            self.head = Node(value)
            self.tail = self.head
        else:
            self.tail.next = Node(value)
            self.tail = self.tail.next
        self.size += 1

    def deq(self):
        if self.size == 0:
            return
        if self.size == 1:
            ret = self.head.data
            self.head = None
            self.tail = None
        else:
            ret = self.head.data
            self.head = self.head.next
        
        self.size -= 1
        return ret
        
    def __str__(self):
        probe = self.head
        ret = ''
        while probe != None:
            ret += f'{probe.data}, '
            probe = probe.next
        return ret[:-2]
```

```python
class Circular_queue:
    def __init__(self, CAPACITY):
        self.CAP = CAPACITY
        self.head = 0
        self.tail = 0
        self.size = 0
        self.arr = [None for _ in range(CAPACITY)]

    def enq(self, value):
        if self.size == self.CAP: # full alr
            return
            
        if self.size == 0:
            self.arr[self.tail] = value
        else:
            self.tail += 1
            if self.tail == self.CAP:
                self.tail = 0
            self.arr[self.tail] = value
        
        self.size += 1

    def deq(self):
        if self.size == 0: # empty alr
            return
            
        if self.size == 1:
            ret = self.arr[self.head]
            self.arr[self.head] = None
            
        else:
            ret = self.arr[self.head]
            self.arr[self.head] = None
            self.head += 1
            if self.head == self.CAP:
                self.head = 0
            
        self.size -= 1
        return ret
    
    def __str__(self):
        return str(self.arr)
```

```python
class Stack:
    def __init__(self):
        self.top = None
        self.size = 0
        
    def push(self, item):
        self.top = Node(item, self.top)
        self.size += 1
    
    def pop(self):
        if self.size == 0:
            return None
        ret = self.top.data
        self.top = self.top.next
        self.size -= 1
        return ret
        
    def peek(self):
        return self.top.data if self.size >= 0 else None
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        ret = ''
        probe = self.top
        for thing in range(self.size):
            ret += f'{probe.data}, '
            probe = probe.next
        return ret[:-2]

    
```

```python
class Linkedlist:
    def __init__(self):
        self.head = None
        self.size = 0
    
    def insert(self, val, pos):
        if pos > self.size: # invalid insert
            return
        if self.size == 0:
            self.head = Node(val)
            
        else: # not empty
            if pos == 0:
                self.head = Node(val, self.head)
            else:
                probe = self.head
                for i in range(pos-1):
                    probe = probe.next
                probe.next = Node(val, probe.next)
                
        self.size += 1
    
    def delete(self, pos): # delete node at index pos
        if pos > self.size: # invalid delete
            return
        if pos == 0:
            ret = self.head.data
            self.head = self.head.next
        else:
            probe = self.head
            for i in range(pos-1):
                probe = probe.next
            # get to the node before the one to be deleted
            ret = probe.next.data
            probe.next = probe.next.next

        self.size -= 1
        return ret

    def bubble_sort(self):
        done = False
        runs = 0
        while not done:
            done = True # toggle it whenever swap is made
            probe = self.head
            for i in range(self.size - 1 - runs):
                if probe.data > probe.next.data: # Need to swap
                    probe.data, probe.next.data = probe.next.data, probe.data
                    done = False
                probe = probe.next
            runs += 1    
            
    def __len__(self):
        return self.size
    
    def __str__(self):
        ret = ''
        probe = self.head
        for thing in range(self.size):
            ret += f'{probe.data}, '
            probe = probe.next
        return ret[:-2]
```

```python
class PQueue:
    def __init__(self):
        self.Initialise()
        
    def Initialise(self):
        self.Front = None
        self.Rear = None
        self.NextFree = 0
        self.ThisPQueue = [Node(None, None, i+1) for i in range(10)]
        self.ThisPQueue[-1].Next = None
    
    def DisplayPQueue(self):
        print('PQ contents')
        print(f"Front, Rear, NextFree: {self.Front}, {self.Rear}, {self.NextFree}")
        print()
        print('Index\t\tData\t\tPriority\tNext')
        for i in range(10):
            print(f'{i}\t\t{self.ThisPQueue[i].Data}\t\t{self.ThisPQueue[i].Priority}\t\t{self.ThisPQueue[i].Next}')
        print()

    
    def PQInsert(self, NewItem, Priority):
        # if empty
        if self.Front == self.Rear == None:
            # get a node
            newnode = self.ThisPQueue[self.NextFree]
            new_idx = self.NextFree
            self.NextFree = self.ThisPQueue[self.NextFree].Next
            # put stuff inside
            newnode.Data = NewItem
            newnode.Priority = Priority
            newnode.Next = None
            # insert into queue
            self.Rear = new_idx
            self.Front = new_idx
        
        # if full
        elif self.NextFree == None:
            print('ERROR: full, cannot insert')
        
        # normal case
        else:
            # get a node
            newnode = self.ThisPQueue[self.NextFree]
            new_idx = self.NextFree
            self.NextFree = self.ThisPQueue[self.NextFree].Next
            # put stuff inside
            newnode.Data = NewItem
            newnode.Priority = Priority
            newnode.Next = None
            # insert into queue
            self.ThisPQueue[self.Rear].Next = new_idx
            self.Rear = new_idx
    
    def PQDelete(self):
        # if empty
        if self.Front == self.Rear == None:
            print('ERROR: empty, cannot delete')
        else:
            # scan through once to get minimum priority
            priorities = []
            for i in self.ThisPQueue:
                if i.Priority != None:
                    priorities.append(i.Priority)
            lowest = min(priorities)
            probe = self.Front
            
            # edge case of removing the front
            if self.ThisPQueue[probe].Priority == lowest:
                # we want to remove the front
                remove = probe
                self.Front = self.ThisPQueue[self.Front].Next

                # clean
                self.ThisPQueue[remove].Data = None
                self.ThisPQueue[remove].Priority = None

                # link to free list
                self.ThisPQueue[remove].Next = self.NextFree

                # set head of free to this
                self.NextFree = remove
                return

            # normal case
            while self.ThisPQueue[self.ThisPQueue[probe].Next].Priority != lowest:
                probe = self.ThisPQueue[probe].Next
                
            # we want to remove the next one
            remove = self.ThisPQueue[probe].Next
            self.ThisPQueue[probe].Next = self.ThisPQueue[self.ThisPQueue[probe].Next].Next
            
            # clean
            self.ThisPQueue[remove].Data = None
            self.ThisPQueue[remove].Priority = None
            
            # link to free list
            self.ThisPQueue[remove].Next = self.NextFree
            
            # set head of free to this
            self.NextFree = remove
```

```python
class Tree:
    def __init__(self, data = None, left = None, right = None):
        self.root = data
        self.left = left
        self.right = right
        
    def insert(self, value):
        if self.root == None:
            self.root = value
        else:
            if value < self.root:
                if self.left == None:
                    self.left = Tree(value)
                else:
                    self.left.insert(value)
            else:
                if self.right == None:
                    self.right = Tree(value)
                else:
                    self.right.insert(value)

    def search(self, value):
        if self.root == None:
            return False
        if value == self.root:
            return True
        elif value < self.root:
            if self.left == None:
                return False
            return self.left.search(value)
        else:
            if self.right == None:
                return False
            return self.right.search(value)
    
    def print_inorder(self):
        if self.left != None:
            self.left.print_inorder()
        print(self.root)
        if self.right != None:
            self.right.print_inorder()
    
    def print_postorder(self):
        if self.left != None:
            self.left.print_postorder()
        if self.right != None:
            self.right.print_postorder()
        print(self.root)
    
    def print_preorder(self):
        print(self.root)
        if self.left != None:
            self.left.print_preorder()
        if self.right != None:
            self.right.print_preorder()
    
    def __len__(self):
        ret = 0
        if self.root != None:
            ret += 1
        if self.left != None:
            ret += len(self.left)
        if self.right != None:
            ret += len(self.right)
        return ret
        
                
```
