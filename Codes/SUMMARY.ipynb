```python
# search algo
def linear(arr, target):
    for i in arr:
        if i == target:
            return True
    return False

def binary(sortedarr, target):
    low, high = 0, len(sortedarr)
    while low < high:
        mid = (low + high)//2
        if target < sortedarr[mid]: # left
            high = mid
        elif target > sortedarr[mid]: # right
            low = mid+1
        else: # same
            return True
    return False

def hashtable(hashtable, target):
    key = hash_function(target)
    
    if hashtable[key] == target:
        return True
    elif hashtable[key] == None: nothing
        return False
    else:
        # linear probe
        key = (key+1)%len(hashtable) # loop this if loop 
        
        # chaining;
        for i in hashtable[key]:
            if i == target:
                return True
        return False
```

```python
# sort algo
def bubble(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

def insert(arr):
    for i in range(len(arr)):
        curr = i
        while curr >= 1 and arr[curr] < arr[curr-1]:
            arr[curr], arr[curr-1] = arr[curr-1], arr[curr]
            curr -= 1
        
def quick(arr, low=0, high=-1):
    if high == -1: # first call of function
        high = len(arr)
    
    if low >= high: # terminal case
        return arr
    else:
        separator = low+1
        for i in range(low+1, high):
            if arr[i] < arr[low]: # less than separator
                arr[i], arr[separator] = arr[separator], arr[i] # swap with big
                separator += 1
        arr[low], arr[separator-1] = arr[separator-1], arr[low] # swap separator in middle
        
        quick(arr, low, separator-1)
        quick(arr, separator, high)

def merge(arr):
    if len(arr) <= 1:
        return arr
    else:
        mid = len(arr)//2
        left = merge(arr[:mid])
        right = merge(arr[mid:])
        
        ret = [] # return array
        while len(left) > 0 and len(right) > 0:
            if left[0] < right[0]:
                ret.append(left.pop(0))
            else: 
                ret.append(right.pop(0))
        ret = ret + left + right
        return ret
```

```python
# data structures
class Node():
    def __init__(self, data, ptr=None):
        self.data = data
        self.ptr = ptr
        
class Stack():
    def __init__(self):
        self.root = None
        
    def push(self, data):
        if self.root == None:
            self.root = Node(data)
        else:
            newNode = Node(data, self.root)
            self.root = newNode
            
    def pop(self):
        if self.root == None:
            print("Error: Stack is empty")
        else:
            popped = self.root
            self.root = popped.ptr
            return popped.data
        
    def __str__(self):
        outp = "Stack: [" # output string
        curr = self.root
        while curr != None:
            outp += f"{curr.data}, "
            curr = curr.ptr
        outp = outp[:-2] + "]"
        return outp
    
class Queue():
    def __init__(self):
        self.head = self.tail = None
        
    def insert(self, data):
        if self.head == None: # empty
            self.head = self.tail = Node(data)
        else:
            newNode = Node(data)
            self.tail.ptr = newNode
            self.tail = newNode
        
    def remove(self):
        if self.head == None: # empty
            print("Error: Queue empty")
        else:
            removed = self.head
            if removed.ptr == None: # end of array
                self.head = self.tail = None
            else:
                self.head = removed.ptr
            return removed.data
        
    def __str__(self):
        outp = "Queue: [" # output string
        curr = self.head
        while curr != None:
            outp += f"{curr.data}, "
            curr = curr.ptr
        outp = outp[:-2] + "]"
        return outp

class CircularQueue():
    def __init__(self, size):
        self.arr = [None] * size
        self.front, self.rear = -1, -1
        
    def isFull(self):
        for i in self.arr:
            if i == None:
                return False
        return True

    def insert(self, data):
        if self.isFull():
            print("Error: Circular queue is full")
        elif self.front == -1: # empty list
            self.front = self.rear = 0
            self.arr[0] = Node(data)
        else:
            self.rear = (self.rear+1)%len(self.arr)
            self.arr[self.rear] = Node(data)
        
    def remove(self):
        if self.front == -1: # empty
            print("Error: Circular queue is empty")
        else:
            ret = self.arr[self.front].data
            self.arr[self.front] = None 
            if self.front == self.rear: # last item
                self.front = self.rear = -1
            else:
                self.front = (self.front+1)%len(self.arr)
            return ret
        
    def __str__(self):
        outp = "Index: \tData:"
        for i in range(len(self.arr)):
            if self.arr[i] == None:
                outp += f"\n{i}: \t-"
            else:
                outp += f"\n{i}: \t{self.arr[i].data}"
        return outp
    
class FreeArray(): # incomplete
    def __init__(self, size):
        self.arr = [] 
        for i in range(size-1):
            self.arr.append(Node(None, i+1))
        self.arr.append(Node(None, -1))
        
        self.head, self.tail = -1, -1
        self.free = 0
        
    def insert(self, data):
        if self.free == -1: # full list
            print("Error: List is full")
        else:
            newNode = self.free
            self.free = self.arr[newNode].ptr # reset self.free
            self.arr[self.tail].ptr = newNode
            self.arr[newNode].data = data
            self.arr[newNode].ptr = -1
            self.tail = newNode
            
            
    def __str__(self):
        outp = f"Head: {self.head} \tTail: {self.tail} \tFree: {self.free}" + \
                "\nIndex: \tData: \tPtr:"
        for i in range(len(self.arr)):
            if self.arr[i] == None:
                outp += f"\n{i}: \t-"
            else:
                outp += f"\n{i}: \t{self.arr[i].data} \t{self.arr[i].ptr}"
        return outp
                
free = FreeArray(5)
free.insert(1)
free.insert(2)
free.insert(3)
free.insert(4)
print(free)
```
